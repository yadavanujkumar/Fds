```python
import networkx as nx
import matplotlib.pyplot as plt

# User input for the number of vertices and edge probability
print("Enter number of vertices:")
n = int(input())
p = float(input("Enter the probability for every node (between 0 and 1):"))

# Generate the random graph
G = nx.erdos_renyi_graph(n, p)

# Check connectivity
is_connected = nx.is_connected(G)

# Find connected components
components = list(nx.connected_components(G))
num_components = len(components)

# Find the largest connected component
largest_component = max(components, key=len)  # Get the component with the maximum size
largest_component_size = len(largest_component)  # Size of the largest component

```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Layout for better visualization

nx.draw(G, pos, with_labels=True,  edge_color="gray", node_size=500)
plt.title(f"Random Graph Connectivity Analysis")
plt.show()

# Print results
print(f"Graph Connected: {is_connected}")
print(f"Number of Connected Components: {num_components}")
print(f"Size of Largest Component: {largest_component_size}")
```
