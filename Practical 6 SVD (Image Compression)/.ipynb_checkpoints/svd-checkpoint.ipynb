```python
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
import numpy as np

# Set figure size for plots
plt.rcParams['figure.figsize'] = [16.8, 8.4]

# Read and process the image
A = imread('download.jpeg')

# Convert the image to grayscale by averaging the RGB channels
X = np.mean(A, axis=-1)

# Display the original grayscale image
plt.imshow(X, cmap='gray')
plt.axis('off')
plt.title('Original Grayscale Image')
plt.show()

# Perform Singular Value Decomposition (SVD)
U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)

# Approximate the image using different numbers of singular values
for j, r in enumerate([5, 20, 100]):
    Xapprox = U[:, :r] @ S[:r, :r] @ VT[:r, :]
    
    plt.figure()
    plt.imshow(Xapprox, cmap='gray')
    plt.axis('off')
    plt.title(f'r = {r} Singular Values')
    plt.show()
```
