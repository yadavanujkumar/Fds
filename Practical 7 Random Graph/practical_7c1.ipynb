```python
import pandas as pd
import networkx as nx
df=pd.read_csv("test.csv")
# Create a graph from the edge list
G = nx.from_pandas_edgelist(df, source='Name', target='Connected')
```

```python
pos=nx.circular_layout(G)
nx.draw(G, with_labels=True, node_color='Aqua', edge_color="gray", node_size=500)
```
