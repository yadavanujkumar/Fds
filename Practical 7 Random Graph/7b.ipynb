```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creating degree distrubition histogram for a random graoh
n=int(input("enter the number of vertices: "))
p=float(input("enter probablity: "))
g=nx.erdos_renyi_graph(n,p)
degrees=[g.degree(node) for node in g.nodes()]
plt.hist(degrees , bins = range(min(degrees), max(degrees) + 1), align = 'left',color = 'coral',edgecolor = 'black')
plt.title("Degree Distribution") 
plt.xlabel("Degree") 
plt.ylabel("Frequency") 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()

```

```python
# Cycles appear when d<1 
n = int(input("Enter the number of vertices: ")) 
d=(float(input("Enter the value of d for cycles to appear(d<1):")))
p = d/n 
g = nx.erdos_renyi_graph(n,p) 
degrees = [g.degree(node) for node in g.nodes()] 
plt.hist(degrees , align = 'left',edgecolor = 'black') 
plt.title("Degree Distribution") 
plt.xlabel("Degree") 
plt.ylabel("Frequency") 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show() 
pos=nx.circular_layout(g) 
nx.draw(g,pos,with_labels=range(1)) 
plt.show() 
```

```python
# Existence of giant components when d>1 
n = int(input("Enter the number of vertices: ")) 
d=(float(input("Enter the value of d:"))) 
p = d/n 
g = nx.erdos_renyi_graph(n,p) 
degrees = [g.degree(node) for node in g.nodes()] 
plt.hist(degrees , align = 'left',edgecolor = 'black') 
plt.title("Degree Distribution") 
plt.xlabel("Degree") 
plt.ylabel("Frequency") 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show() 
pos=nx.circular_layout(g) 
nx.draw(g,pos,with_labels=1) 
plt.show() 
```

```python
# Existence of giant components with isolated vertices 
n = int(input("Enter the number of vertices: ")) 
p = 0.5*((np.log(n))/n) 
g = nx.erdos_renyi_graph(n,p) 
degrees = [g.degree(node) for node in g.nodes()] 
plt.hist(degrees, align = 'left',edgecolor = 'black') 
plt.title("Degree Distribution") 
plt.xlabel("Degree") 
plt.ylabel("Frequency") 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show() 
pos=nx.circular_layout(g) 
nx.draw(g,pos,with_labels=1) 
plt.show() 
```
